```python
#Paul Phillips
import sklearn #importing libraries
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pandas as pd

location = 'preptrain.csv' #obtaining location of data file
data = pd.read_csv(location)
data.head()

X = data.drop('SalePrice', axis = 1) #SalePrice is the target so it is dropped from X
y = data.SalePrice #y is the target
#data is split along 80 for training and 20 for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8945, shuffle = True)

gbRegression = GradientBoostingRegressor()
gbRegression.fit(X_train, y_train) #the gradient boost regressor is trained on the training set
gReg = gbRegression.predict(X_test) #gradient boost regressor makes predictions on the test set
r2Result = r2_score(y_test,gReg)

print("R2 score will be used to evalue this Gradient Boost regression model")
print("Test set is 20% of the data")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.2) #learning rate raised from 0.1 to 0.2
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("With the learning rate doubled to 0.2")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.8) #learning rate raised to 0.8
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("With the learning rate raised to 0.8")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.05) #learning rate reduced to 0.05
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("With the learning rate lowered to 0.05")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

print("Learning rate of 0.1 gives the best results so far")
print(" ")


gbRegression = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 300) #raises number of estimators to 300
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("The estimators are rasied to 300 and the learning rate is reset to 0.1")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 800) #the estimator numbers are raised to 800
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("With good results from increasing estimators to 300, here they are increased to 800")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 1100) #estimators at 1100
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)
bestReg = gbRegression

print("At 1100 estimators")
print("When tested, the model's r2 score is:", r2Result)
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 1100, max_depth = 5) #estimator depth is raised to 5
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("Estimator depth is now raised from 3 to 5")
print("When tested, the model's r2 score is:", r2Result)
print(" ")
print(" ")

gbRegression = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 1100, max_depth = 2) #estimator depth is lowered to 2
gbRegression.fit(X_train, y_train)
gReg = gbRegression.predict(X_test)
r2Result = r2_score(y_test,gReg)

print("Estimator depth is now lowered to 2")
print("When tested, the model's r2 score is:", r2Result)
print(" ")
print("The 1100 estimator model without adjustments to estimator depth is the best model")

import pickle
import os
pickle.dump(bestReg, open(os.path.join('/content/', 'gboostModel.pkl'), 'wb'), protocol=4)
```
